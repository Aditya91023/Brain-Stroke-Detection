```python
import zipfile
import os

with zipfile.ZipFile("Brain_Data_Organised.zip", "r") as zip_ref:
    zip_ref.extractall("/content/")
```

```python
# If using Colab, upload the extracted "Brain_Data_Organised" folder to `/content/`
dataset_path = "/content/Brain_Data_Organised"
stroke_dir = os.path.join(dataset_path, "Stroke")
normal_dir = os.path.join(dataset_path, "Normal")
```

```python
def load_images_from_folder(folder, label, img_size=(224, 224)):
    images = []
    labels = []
    for filename in os.listdir(folder):
        path = os.path.join(folder, filename)
        try:
            img = cv2.imread(path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, img_size)
            images.append(img)
            labels.append(label)
        except:
            continue
    return images, labels

stroke_images, stroke_labels = load_images_from_folder(stroke_dir, 1)
normal_images, normal_labels = load_images_from_folder(normal_dir, 0)

X = np.array(stroke_images + normal_images)
y = np.array(stroke_labels + normal_labels)

print("Data shape:", X.shape)
print("Label distribution:", np.bincount(y))
```

```python
X = X / 255.0  # Normalize
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
datagen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=True,
)

train_generator = datagen.flow(X_train, y_train, batch_size=16)
```

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D(),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Binary classification
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.AUC()])
```

```python
callbacks = [
    EarlyStopping(patience=5, restore_best_weights=True),
    ModelCheckpoint("best_model.h5", save_best_only=True)
]

history = model.fit(
    train_generator,
    validation_data=(X_test, y_test),
    epochs=15,
    callbacks=callbacks
)
```

```python
loss, acc, auc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {acc:.2f}, AUC: {auc:.2f}")

# Confusion Matrix
y_pred = (model.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
def plot_history(hist):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    plt.plot(hist.history['accuracy'], label='Train Acc')
    plt.plot(hist.history['val_accuracy'], label='Val Acc')
    plt.title('Accuracy')
    plt.legend()

    plt.subplot(1, 3, 2)
    plt.plot(hist.history['loss'], label='Train Loss')
    plt.plot(hist.history['val_loss'], label='Val Loss')
    plt.title('Loss')
    plt.legend()

    plt.subplot(1, 3, 3)
    plt.plot(hist.history['auc'], label='Train AUC')
    plt.plot(hist.history['val_auc'], label='Val AUC')
    plt.title('AUC')
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_history(history)
```

```python
import gradio as gr
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

# === Load trained model ===
model = load_model("stroke_classifier_model.keras")  # or .h5
print("✅ Model loaded!")

# === Prediction function ===
def predict_stroke(img):
    # Resize and normalize
    img_resized = img.resize((224, 224))
    img_array = image.img_to_array(img_resized) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Predict
    prediction = model.predict(img_array)[0][0]
    label = "🧠 Stroke" if prediction > 0.5 else "✅ Normal"
    confidence = prediction if prediction > 0.5 else 1 - prediction
    confidence_percent = f"{confidence * 100:.2f}%"

    return label, confidence_percent

# === Gradio UI ===
with gr.Blocks(theme=gr.themes.Base()) as demo:
    gr.Markdown("## 🧠 Brain Stroke Detection from CT Image")
    gr.Markdown("Upload a CT image to detect if it's a stroke or normal brain scan.")

    with gr.Row():
        with gr.Column():
            img_input = gr.Image(type="pil", label="Upload CT Image")
            predict_btn = gr.Button("🔍 Predict")

        with gr.Column():
            label_output = gr.Label(label="Prediction")
            conf_output = gr.Label(label="Confidence")

    predict_btn.click(fn=predict_stroke, inputs=img_input, outputs=[label_output, conf_output])

# === Launch UI ===
demo.launch(share=True)  # use share=True to get public link in Colab
```
